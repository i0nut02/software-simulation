```python
import pandas as pd
```

```python
file_path = "./software-simulation/logfile.txt"
df = pd.read_csv(file_path, delimiter=';')
df['value'] = pd.to_numeric(df['value'], errors='coerce')

filtered_row = df[df['cmd'] == 7]
server_num = filtered_row.iloc[0]['stream'].split("-")[0].strip()
print(server_num)

df.head(305)
```

```python
grouped_df = df.groupby(['stream', 'cmd'])
```

```python
count_per_group = grouped_df.size().reset_index(name='count')
count_per_group_sorted = count_per_group.sort_values(ascending=False, by = 'count')
pd.set_option('display.max_rows', None)
count_per_group_sorted.head(305)
```

```python
filter_df = count_per_group_sorted[(count_per_group_sorted['stream'] != 'orchestrator-' + str(server_num)) &
                                    count_per_group_sorted['stream'].str.startswith('orchestrator-')]

mean_req_per_client = filter_df['count'].mean() / 2
mean_req_per_second_per_client = mean_req_per_client / 10 / 60 
print(mean_req_per_client, mean_req_per_second_per_client)
filter_df.head(300)
```

```python
filter_df2 = count_per_group_sorted[(count_per_group_sorted['stream'] != 'orchestrator-' + str(server_num)) &
                                    count_per_group_sorted['stream'].str.endswith('orchestrator')]
filter_df2.head(300)
```
